```python
#import libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model

df = pd.read_csv('honeyproduction.csv')
```

```python
#Get a sense of the dataset

print(df.head())
```

```python
#Calculate average production per year 
prod_per_year = df.groupby('year').totalprod.mean().reset_index()
```

```python
#Create X and y variables for linear regression model. X is the year, y is the total production.
X = prod_per_year['year']
X = X.values.reshape(-1,1)
y = prod_per_year['totalprod']

```

```python
#display data

plt.scatter(X,y)
```

```python
#create and fit the linear regression model using scikit learn
#create the model

regr = linear_model.LinearRegression()

#fit the model
regr.fit(X,y)

coef = (regr.coef_[0])

y_predict = regr.predict(X)
plt.scatter(X, y)
plt.plot(X, y_predict)
plt.show()
```

```python
#predict what honey production will be in 2050

X_future = np.array(range(2013,2051))

print(X_future)
X_future = X_future.reshape(-1,1)

print(X_future)
```

```python
future_predict = regr.predict(X_future)

plt.plot(X_future, future_predict)
plt.show()
```
